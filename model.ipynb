```python
from langchain.llms import Ollama
llm = Ollama(model="llama3:latest")
test_questions = [
    "교통비 절감에 좋은 카드는?",
    "배달앱 혜택이 많은 신용카드는?",
    "여행할 때 공항라운지 혜택이 있는 카드 추천해줘.",
    "쿠팡 할인되는 카드 알려줘.",
    "가장 높은 캐시백 제공하는 카드 알려줘."
]

for question in test_questions:
    response = llm(question)
    print(f"🚀 질문: {question}\n➡ 답변: {response}\n")

```
